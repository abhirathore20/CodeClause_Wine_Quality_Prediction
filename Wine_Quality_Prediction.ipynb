**1. Importing the Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

**2. Data Collection**

```python
# loading the dataset to a Pandas DataFrame
wine_dataset = pd.read_csv("/content/winequality-red.csv")
```

```python
# number of rows and columns in the dataset
wine_dataset.shape
```

```python
# first 5 rows of the dataset
wine_dataset.head()
```

```python
# checking for missing values
wine_dataset.isnull().sum()
```

**3. Data Analysis and Visualization**

```python
# statistical measures of the dataset
wine_dataset.describe()
```

```python
# display histogram
wine_dataset.hist(bins=25,figsize=(10,10))
plt.show()
```

```python
# Alcohol vs Quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y = 'alcohol', data = wine_dataset)
```

**3. Correlation**

1.Positive Correlation

2.Negative Correlation

```python
correlation = wine_dataset.corr()
```

```python
# constructing a heatmap to understand the correlation between the columns
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt = '.1f', annot = True, cmap = 'Greens')
```

**4. Data Preprocessing**

```python
 # separate the data and Label
X = wine_dataset.drop('quality',axis=1)
```

```python
print(X)
```

**5. Label Binarizaton**

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
print(Y)
```

**6. Train & Test Split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

**7. Model Train:**

Random Forest Classifier  

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

**8. Model Evaluation**

Accuracy Score

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy : ', test_data_accuracy)
```

**9. Building a Predictive System**

```python
# Positive Result

input_data=(7.5,0.520,0.16,1.9,0.085,12.0,35.0,0.99680,3.38,0.62,9.5)

#changing the input data to a numpy array

input_data_as_np=np.asarray(input_data)

#reshape the data

input_data_reshaped=input_data_as_np.reshape(1,-1)

prediction=model.predict(input_data_reshaped)

print(prediction)

if(prediction[0]==1):
    print('Good Quality Wine')
else:
    print('Bad Quality Wine')
```

```python
# Negative Result

input_data=(7.3,0.305,0.39,1.2,0.059,7.0,11.0,0.99331,3.29,0.52,11.5)

#changing the input data to a numpy array

input_data_as_np=np.asarray(input_data)

#reshape the data

input_data_reshaped=input_data_as_np.reshape(1,-1)

prediction=model.predict(input_data_reshaped)

print(prediction)

if(prediction[0]==1):
    print('Good Quality Wine')
else:
    print('Bad Quality Wine')
```
